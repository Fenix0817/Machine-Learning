# Seaborn Tutorial

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
# Using tips dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Dist Plot or Distribution plot
# KDE: Kernel Density Estimation Plot
sns.distplot(tips['total_bill'],kde=False, bins=20)
```

```python
# Combine two different distplots
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='hex')
```

```python
# Plots regression line
sns.jointplot(x='total_bill',y='tip',data=tips, kind='reg')
```

```python
# Density of points
# KDE: Kernel Density Estimation Plot
sns.jointplot(x='total_bill',y='tip',data=tips, kind='kde')
```

```python
# Pair Plot: Pairwise relationships for all columns in data
# Does Jointplot for every single combination in dataframe (numerical values only)
sns.pairplot(tips)
```

```python
# Hue to add categorical columns like SEX
sns.pairplot(tips, hue='sex')
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
# Rug Plot
# Draws a | mark for every sample in data
sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
tips.head()
```

# Categorical Plots

```python
# Categorical Plots

# Bar Plot
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
```

```python
# Count number of occurences
sns.countplot(x='sex', data=tips)
```

```python
# Box Plot
sns.boxplot(x='day',y='total_bill',data=tips, hue='smoker')
```

```python
# Violin Plot
sns.violinplot(x='day',y='total_bill',data=tips, hue='sex', split=True)
```

```python
# Strip Plot: draws a Scatter Plot
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
# Add Jitter to see density of points a little better
sns.stripplot(x='day',y='total_bill',data=tips, jitter=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=True, hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=True, hue='sex', split=True)
```

```python
# Swarm Plot: Points adjusted so that they don't overlap
# Drawback: does not scales well to large data sometimes
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```

```python
# Factor Plot
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='violin')
```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='strip')
```

# Matrix Plots

```python
# Matrix Plots
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
# Heat Maps
# Data should be in Matrix Form for this
tc = tips.corr()
tc
```

```python
sns.heatmap(tc)
```

```python
# Show values in heatmap
sns.heatmap(tc, annot=True)
```

```python
sns.heatmap(tc, annot=True, cmap='coolwarm')
```

```python
# Set Index and Convert Data to Matrix Form
fp = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
fp
```

```python
# Darker the color, more the number value
sns.heatmap(fp)
```

```python
sns.heatmap(fp, cmap='magma')
```

```python
sns.heatmap(fp,cmap='magma',linecolor='white', linewidths=1)
```

```python
sns.heatmap(fp,cmap='coolwarm',linecolor='white', linewidths=1)
```

# Grids

```python
# Iris Dataset
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
# Pair Plot
sns.pairplot(iris)
```

```python
# PairGrid
import matplotlib.pyplot as plt

p = sns.PairGrid(iris)
p.map(plt.scatter)
```

```python
# Diagonal Plot
# Choose what type plots we want on grid
p = sns.PairGrid(iris)
p.map_diag(sns.distplot)
p.map_upper(plt.scatter)
p.map_lower(sns.kdeplot)
```

```python
tips.head()
```

```python
# Facet Grid
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.distplot,'total_bill')

# So, top row: Smokers, Bottom row: Non-Smokers
# 1st Column: time: Lunch, 2nd Column, time:Dinner
# x-axis: total_bill
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(plt.scatter,'total_bill', 'tip')
```
