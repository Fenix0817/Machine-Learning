# Pandas Data Visualization Tutorial

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df1 = pd.read_csv('df1',index_col=0)
```

```python
df1.head()
```

```python
df2 = pd.read_csv('df2')
```

```python
df2.head()
```

```python
# Create a Histogram Plot
df1['A'].hist(bins=20)
```

```python
df1['A'].plot(kind='hist', bins=20)
```

```python
df1['A'].plot.hist()
```

```python
# Plot the area for whole data
df2.plot.area()
```

```python
df2.plot.area(alpha=0.3)
```

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
```

```python
df1['A'].plot.hist(bins=20)
```

```python
# Linea Plot
df1.plot.line(x=df1.index, y='B')
```

```python
df1.plot.line(x=df1.index, y='B', figsize=(12,3))
```

```python
# Scatter Plot
df1.plot.scatter(x='A',y='B')
```

```python
df1.plot.scatter(x='A',y='B',c='C')
```

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

```python
df1.plot.scatter(x='A', y='B', s=df1['C']*100)
```

# Box Plots

```python
df2.plot.box()
```

```python
df = pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])
```

```python
df.plot.hexbin(x='a',y='b')
```

```python
df.plot.hexbin(x='a',y='b', gridsize=25)
```

```python
df.plot.hexbin(x='a',y='b', gridsize=25, cmap='coolwarm')
```
